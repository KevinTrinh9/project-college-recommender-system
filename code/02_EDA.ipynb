# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/data_cleaned.csv')
```

```python
data
```

```python
data.describe()
```

```python
# filter out "Not Available" values from "SAT_AVG" column and plot histogram
sat_scores = data[data['SAT_AVG'] != 'Not Available']['SAT_AVG'].astype(float)
plt.figure(figsize=(10, 6))
sns.histplot(sat_scores, kde=True, color='skyblue')
plt.axvline(sat_scores.mean(), color='red')
plt.title('Distribution of Average SAT Scores')
plt.xlabel('Average SAT Score')
plt.ylabel('Frequency');
```

Interpretation: Slightly right-skewed distribution, with an average overall SAT score between 1100 and 1200 across all colleges.

```python
# filter out "Not Available" values from "ADM_RATE" column and plot histogram
admission_rates = data[data['ADM_RATE'] != 'Not Available']['ADM_RATE'].astype(float)
plt.figure(figsize=(10, 6))
sns.histplot(admission_rates, kde=True, color='lightgreen')
plt.axvline(admission_rates.mean(), color='red')
plt.title('Distribution of Admission Rates')
plt.xlabel('Admission Rate')
plt.ylabel('Frequency');
```

Interpretation: Left-skewed distribution with an average overall admission rate of about 70% across all colleges.

```python
data['PREDDEG'].value_counts()
```

```python
# count amount of each degree/certificate level in the "PREDDEG" column and create bar plot
degree_counts = data['PREDDEG'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=degree_counts.index, y=degree_counts.values, color='plum')
plt.title('Distribution of Institutions by Predominant Degree/Certificate Awarded')
plt.xlabel('Degree/Certificate Level')
plt.ylabel('Frequency');
```

Interpretation: A majority of institutions award predominantly Associate's (2) or Bachelor's (3) degrees, but there are almost just as many institutions that award predominantly certificates (1). There are very few institutions that award only graduate (4) degrees.

```python
data['HIGHDEG'].value_counts()
```

```python
# count amount of the highest degree/certificate levels in the "HIGHDEG" column and create bar plot
high_degree_counts = data['HIGHDEG'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=high_degree_counts.index, y=high_degree_counts.values, color='teal')
plt.title('Distribution of Institutions by Highest Degree/Certificate Awarded')
plt.xlabel('Highest Degree/Certificate Level')
plt.ylabel('Frequency');
```

Interpretation: A majority of institutions award Associate's (2), Bachelor's (3), or graduate (4) level degrees as their highest level, but there are also many institutions that award only certificates (1). There are very few institutions that are non-degree/certificate-granting.

```python
data['STABBR'].value_counts()
```

```python
# count number of colleges per state in the "STABBR" column and create bar plot
college_counts_per_state = data['STABBR'].value_counts()
plt.figure(figsize=(8, 15))
sns.barplot(x=college_counts_per_state.values, y=college_counts_per_state.index, color='coral')
plt.title('Number of Colleges per State/U.S. Territory')
plt.xlabel('Number of Colleges')
plt.ylabel('State/U.S. Territory')
plt.savefig('colleges_per_state.png');
```

Interpretation: The states/U.S. territories with the largest number of colleges are California, New York, and Texas.

```python
data['CURROPER'].value_counts()
```

```python
# Count the frequency of each operational status in the 'CURROPER' column and create bar plot
operational_counts = data['CURROPER'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=operational_counts.index, y=operational_counts.values)
plt.title('Distribution of Institutions by Operational Status')
plt.xlabel('Operational Status')
plt.ylabel('Frequency');
```

Interpretation: Most institutions in the dataset are currently operational.

```python
# create heatmap of numerical features
numerical_data = data.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_data.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, annot_kws={"size": 10})
plt.title('Correlation Heatmap of Numerical Features');
```

```python
# create cluster map of "PCIP" columns
pcip_data = data.filter(like='PCIP')
pcip_corr = pcip_data.corr()
plt.figure(figsize=(12, 10))
sns.clustermap(pcip_corr, cmap='coolwarm', linewidths=0.5)
plt.title('Cluster Map of Fields of Study');
```
