# Data Cleaning

```python
import pandas as pd
```

```python
data = pd.read_csv('data/MERGED2022_23_PP.csv', low_memory=False)
```

```python
data.head()
```

```python
# select specific columns
selected_columns = ['INSTNM', 'CITY', 'STABBR', 'INSTURL', 'PREDDEG', 'HIGHDEG', 'ADM_RATE', 'SAT_AVG', 'CURROPER', 'ADDR']
columns_to_keep = []
columns_to_keep.extend(selected_columns)
for column_name in data.columns:
    if column_name.startswith('PCIP'):
        columns_to_keep.append(column_name)
data = data[columns_to_keep]
```

```python
# check for duplicates in df
data.duplicated().sum()
```

```python
# count missing values in df
data.isna().sum()
```

```python
# replace missing values in specific columns with "Not Available"
columns_to_fill = ['INSTURL', 'ADM_RATE', 'SAT_AVG', 'ADDR']
replacement_value = "Not Available"
data.fillna({col: replacement_value for col in columns_to_fill}, inplace=True)
```
